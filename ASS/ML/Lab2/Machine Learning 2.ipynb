# 3) Exploring relationship between variables by Scatter Plot and Cross tabulation

```python
home1  = pd.read_csv("home1.csv", names=['rooms','rent','sqft'])
home1.head()
```

```python
home2 = pd.read_csv("home2.csv")
home2.head()
```

# Scatter plot

```python
plt.scatter(home1.rooms, home1.rent, marker='*')
plt.xlabel("Rooms")
plt.ylabel("Rent")
plt.show()
```

```python
plt.scatter(home2.rooms, home2.rent, marker='*')
plt.xlabel("Rooms")
plt.ylabel("Rent")
plt.show()
```

```python
plt.scatter(home1.rooms, home1.sqft, marker='+')
plt.xlabel("Rooms", fontweight='bold')
plt.ylabel("Square Feet", fontweight='bold')
plt.show()
```

```python
plt.scatter(home2.rooms, home2.sqft, marker='+')
plt.xlabel("Rooms", fontweight='bold')
plt.ylabel("Square Feet", fontweight='bold')
plt.show()
```

```python
 plt.scatter(home1.rent, home1.sqft, marker='o')
plt.xlabel("Rent", fontweight='bold')
plt.ylabel("Square Feet", fontweight='bold')
plt.show()
```

```python
plt.scatter(home2.rent, home2.sqft, marker='o')
plt.xlabel("Rent", fontweight='bold')
plt.ylabel("Square Feet", fontweight='bold')
plt.show()
```

# Cross Tablulation

```python
pd.crosstab(home1.rooms, home1.rent)
```

```python
pd.crosstab(home2.rooms, home2.rent)
```

```python
pd.crosstab(home1.rooms, home1.sqft)
```

```python
pd.crosstab(home2.rooms, home2.sqft)
```

```python
pd.crosstab(home1.rent, home1.sqft)
```

```python
pd.crosstab(home2.rent, home2.sqft)
```
